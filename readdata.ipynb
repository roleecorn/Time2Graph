```python
results_str = """
result: accu 0.9541, prec 0.9742, recall 0.9538, f1 0.9639
result: accu 0.9595, prec 0.9827, recall 0.9538, f1 0.9680
result: accu 0.9649, prec 0.9870, recall 0.9580, f1 0.9723
result: accu 0.9514, prec 0.9741, recall 0.9496, f1 0.9617
result: accu 0.9595, prec 0.9785, recall 0.9580, f1 0.9682
result: accu 0.9568, prec 0.9744, recall 0.9580, f1 0.9661
result: accu 0.9622, prec 0.9870, recall 0.9538, f1 0.9701
result: accu 0.9541, prec 0.9784, recall 0.9496, f1 0.9638
result: accu 0.9622, prec 0.9828, recall 0.9580, f1 0.9702
result: accu 0.9595, prec 0.9869, recall 0.9496, f1 0.9679
""".strip().split('\n')

results = {}
for result in results_str:
    metrics = result.replace('result: ', '').split(', ')
    for metric in metrics:
        name, value = metric.split(' ')
        if name not in results:
            results[name] = []
        results[name].append(float(value))
```

```python
import numpy as np

for name, values in results.items():
    avg = np.mean(values)
    max_val = np.max(values)
    min_val = np.min(values)
    variance = np.var(values)

    print(f"For {name}:")
    print(f"  Average: {avg}")
    print(f"  Max: {max_val}")
    print(f"  Min: {min_val}")
    print(f"  Variance: {variance}")
    print()
```
