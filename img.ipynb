```python
import pandas as pd
import matplotlib.pyplot as plt

#  tsv 
df = pd.read_csv('output.tsv', sep='\t', header=None)

# 
print(df.head())

# 
positive_sample = df[df[0] == 1].iloc[0, 1:]
negative_sample = df[df[0] == 0].iloc[0, 1:]

# 
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(positive_sample)
plt.title('Positive Sample')
plt.xlabel('Time')
plt.ylabel('Normalized Power')

plt.subplot(1, 2, 2)
plt.plot(negative_sample)
plt.title('Negative Sample')
plt.xlabel('Time')
plt.ylabel('Normalized Power')

plt.tight_layout()
plt.show()
```
